```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
data = pd.read_csv('../data/fandango_score_comparison.csv')
```

```python
#Add Vertical & Horizontal Marker Lines
data.plot.scatter(x='Metacritic', y='Metacritic_User')
plt.axhline(5, color='k')
plt.axvline(50, color='k')
```

```python
#Add Text
data.plot.scatter(x='Metacritic', y='Metacritic_User')
plt.text(10, 9, 'Popcorn Pleasers')
plt.text(70, 4, 'Snob Favourites')
plt.axhline(5, color='k')
plt.axvline(50, color='k')
```

```python
#Use the annotate() method
data.plot.scatter(x='Metacritic', y='Metacritic_User')
plt.axhline(5, color='k')
plt.axvline(50, color='k')
plt.annotate('Paul Blart 2', xy=(13,2.4), xytext=(5,4), arrowprops=dict(facecolor='r', shrink=0.5))
plt.text(10,9, 'Popcorn Pleasers')
plt.text(70,4, 'Snob Favourites')
```
